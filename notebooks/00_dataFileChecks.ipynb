```python
import os
import shutil
print(os.getcwd())
```

```python

# path
base_folder = "../data/nili_fossae_full_collection"
# base_folder = "/media/samuil/408804E68804DBF2/nili_fossae_full_collection"

# loop through all files in the directory
for filename in os.listdir(base_folder):
    if not os.path.isfile(os.path.join(base_folder, filename)):
        continue

    # observation ID
    # format: frt0000b573_07_if166j_mtr3.img
    obs_id = filename.split("_")[0].lower()  # lowercase just in case

    # mmake a folder for this observation ID if it doesn't exist
    obs_folder = os.path.join(base_folder, obs_id)
    os.makedirs(obs_folder, exist_ok=True)

    # move file into the corresponding folder
    src = os.path.join(base_folder, filename)
    dst = os.path.join(obs_folder, filename)
    shutil.move(src, dst)

print("Sorting complete.")
```

```python
# text file with the observation ids
# text_file_path = "/home/samuil/Downloads/crism_product_ids.txt"
text_file_path = "../data/crism_product_ids.txt"


# observation IDs from the text file
with open(text_file_path, "r") as f:
    expected_obs_ids = [line.strip().split("_")[0].lower() for line in f if line.strip()]

# get all folder names currently in the base folder
existing_obs_ids = set(
    name.lower() for name in os.listdir(base_folder)
    if os.path.isdir(os.path.join(base_folder, name))
)

# find missing observation IDs
missing_obs_ids = [obs_id for obs_id in expected_obs_ids if obs_id not in existing_obs_ids]

# display results
print("Missing Observation IDs:")
print(len(missing_obs_ids))
for obs in missing_obs_ids:
    print(obs)
```

```python
# cheack if files are missing from the folders
i = 0
for folder in os.listdir(base_folder):
    folder_path = os.path.join(base_folder, folder)
    if not os.path.isdir(folder_path):
        continue

    expected_extensions = {".img", ".lbl", ".hdr"}
    found_extensions = set()

    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if file.lower().startswith(folder) and os.path.isfile(file_path):
            _, ext = os.path.splitext(file.lower())
            found_extensions.add(ext)

    i = i+1
    missing = expected_extensions - found_extensions
    if missing:
        print(f"[WARNING] {i}. Folder '{folder}' is missing: {', '.join(missing)}")
    else:
        print(f"[OK] {i}. Folder '{folder}' has all required files.")
```
